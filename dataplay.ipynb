```python
import pandas as pd

df = pd.read_csv("us-counties-5_20.csv")

df['date'] = pd.to_datetime(df['date'])
cobb_df = df[df['county'] == 'Cobb']
cobb_df.set_index("date" , inplace=True)
cobb_df['cases'].plot()
```

```python
cobb_df["cases"].rolling(7).mean().plot()
```

```python
graph_df = pd.DataFrame()
for county in df['county'].unique()[:30]:
    print(county)
    county_df = df.copy()[df['county'] == county]
    county_df.set_index('county', inplace = True)
    county_df.sort_index(inplace = True)
    county_df['rolling7avg'] = county_df['cases'].rolling(7).mean()
    
    if graph_df.empty:
        graph_df = county_df[[ f'{county}_rolling7avg']]
    else:
        graph_df = graph_df.join(county_df[f'{county}_rolling7avg'])
```
