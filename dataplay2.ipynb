```python
#dc  = date change
dc =  pd.DataFrame()

for name, group in df.groupby("state"):
    if dc.empty:
        dc  = group.set_index("date")[["cases"]].rename(columns={'cases':name})
    else:
        dc  = dc.join(group.set_index("date"))[["cases"]].rename(columns={'cases':name})
            
dc.head()
```
